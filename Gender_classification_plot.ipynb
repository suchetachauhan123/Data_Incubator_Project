```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import sys
from PIL import Image

plt.figure(dpi=300)
lw = 2

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 12}

plt.rc('font', **font)
```

# Problem Statement 1: Develop a CNN-based model for Predicting patients with Schizophrenia using MRI images.
#### CNN results after applying 72-fold cross validation technique. The model performance in terms of AUC is reported i.e. 0.5759.


```python
x = np.loadtxt('patient_fpr.txt')
y = np.loadtxt('patient_tpr.txt')

auc1 = np.trapz(y, x)

f, ax1 = plt.subplots(1, 1, sharey=False)

ax1.plot(x, y, color='b',label="AUC= %.4f"%auc1)

ax1.set_title('ROC for CNN Model on full dataset', fontweight='bold', fontsize=15)
ax1.set_ylabel('True Positive Rate', fontweight='bold', fontsize=15)
ax1.set_xlabel('False Positive Rate', fontweight='bold', fontsize=15)
ax1.legend(loc='lower right', ncol=1, prop={'size': 15, 'weight':'bold'})
#ax2.set_xlim(0.3,1)
#ax1.set_ylim(0.8, 1)
f.savefig('auc_patient_cnn.png',format="png",dpi = 300)
f.show()
```

# Problem Statement 2: Develop a CNN-based model to predict the gender of a person using MRI images.
#### CNN results after applying 73-fold cross validation technique.  The model performance in terms of AUC is reported i.e. 0.5156.



```python
x = np.loadtxt('gender_fpr.txt')
y = np.loadtxt('gender_tpr.txt')

auc1 = np.trapz(y, x)

f, ax1 = plt.subplots(1, 1, sharey=False)

ax1.plot(x, y, color='b',label="AUC= %.4f"%auc1)

ax1.set_title('ROC for CNN Model on full dataset \nfor Gender classification', fontweight='bold', fontsize=15)
ax1.set_ylabel('True Positive Rate', fontweight='bold', fontsize=15)
ax1.set_xlabel('False Positive Rate', fontweight='bold', fontsize=15)
ax1.legend(loc='lower right', ncol=1, prop={'size': 15, 'weight':'bold'})
#ax2.set_xlim(0.3,1)
#ax1.set_ylim(0.8, 1)
f.savefig('auc_gender_cnn.png',format="png",dpi = 300)
f.show()
```
